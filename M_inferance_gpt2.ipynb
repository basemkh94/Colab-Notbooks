```python
#consolidated code 

from transformers import pipeline, AutoModelWithLMHead, AutoTokenizer
import torch 

model_path = '/content/drive/MyDrive/models_nlp/mod_v1.pth'
saved_model = torch.load(model_path)
tokenizer = AutoTokenizer.from_pretrained("gpt2")
generator = pipeline('text-generation', model = saved_model , tokenizer = tokenizer)
```

```python
def paraphrase(input_sentance):
  return generator('<s>'+input_sentance+ '</s>>>>><p>')
paraphrase("hi, how are you?")
```
